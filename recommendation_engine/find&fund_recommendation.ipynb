```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pickle
import json

# Load data
researchers_df = pd.read_csv('researchers_clean.csv')
grants_df = pd.read_csv('grants_final_cleaned.csv')

# Fill missing values
for col in researchers_df.columns:
   if researchers_df[col].dtype == 'object':
       researchers_df[col] = researchers_df[col].fillna('Unknown')
   else:
       researchers_df[col] = researchers_df[col].fillna(0)

grants_df['close_date'] = pd.to_datetime(grants_df['close_date'], errors='coerce')
grants_df['close_date'] = grants_df['close_date'].fillna(pd.to_datetime('2024-01-01'))

for col in grants_df.columns:
   if grants_df[col].dtype == 'object':
       grants_df[col] = grants_df[col].fillna('Unknown')
   elif pd.api.types.is_numeric_dtype(grants_df[col]):
       grants_df[col] = grants_df[col].fillna(0)
   elif pd.api.types.is_datetime64_any_dtype(grants_df[col]):
       grants_df[col] = grants_df[col].fillna(pd.to_datetime('2024-01-01'))
   else:
       grants_df[col] = grants_df[col].fillna('Unknown')

def create_researcher_text(row):
   text_parts = [
       str(row['concept_1']),
       str(row['concept_2']),
       str(row['top_work_1_keywords']),
       str(row['top_work_2_keywords']),
       str(row['top_work_3_keywords']),
       str(row['top_work_4_keywords']),
       str(row['top_work_5_keywords'])
   ]
   return ' '.join(text_parts)

def create_grant_text(row):
   text_parts = [
       str(row['opportunity_title']),
       str(row['funding_categories']),
       str(row['summary_description']),
       str(row['eligibility_description'])
   ]
   return ' '.join(text_parts)

# Create text features
researcher_text = researchers_df.apply(create_researcher_text, axis=1)
grant_text = grants_df.apply(create_grant_text, axis=1)

# Combine and vectorize text
tfidf = TfidfVectorizer(stop_words='english', max_features=5000)
combined_text = pd.concat([researcher_text, grant_text])
tfidf_matrix = tfidf.fit_transform(combined_text)

# Split back to researcher and grant features
n_researchers = len(researcher_text)
researcher_features = tfidf_matrix[:n_researchers]
grant_features = tfidf_matrix[n_researchers:]

# Convert to dense arrays
researcher_features_dense = np.asarray(researcher_features.todense())
grant_features_dense = np.asarray(grant_features.todense())

# Save model components
np.save('researcher_features.npy', researcher_features_dense)
np.save('grant_features.npy', grant_features_dense)

researcher_ids = researchers_df['researcher_id'].tolist()
grant_ids = grants_df['opportunity_id'].tolist()

with open('researcher_ids.json', 'w') as f:
   json.dump(researcher_ids, f)
with open('grant_ids.json', 'w') as f:
   json.dump(grant_ids, f)
with open('tfidf_vectorizer.pkl', 'wb') as f:
   pickle.dump(tfidf, f)

def get_recommendations(researcher_id, researcher_features, grant_features, researcher_ids, grant_ids, top_n=5):
   researcher_idx = researcher_ids.index(researcher_id)
   similarities = cosine_similarity(
       researcher_features[researcher_idx].reshape(1, -1),
       grant_features
   )[0]

   top_indices = similarities.argsort()[-top_n:][::-1]
   return [
       {
           'grant_id': grant_ids[idx],
           'similarity_score': float(similarities[idx])
       }
       for idx in top_indices
   ]

# Test recommendations
sample_recommendations = get_recommendations(
   researcher_ids[0],
   researcher_features_dense,
   grant_features_dense,
   researcher_ids,
   grant_ids
)
print("Sample recommendations:", sample_recommendations)

# Print shapes to verify
print("\nFeature matrix shapes:")
print(f"Researcher features: {researcher_features_dense.shape}")
print(f"Grant features: {grant_features_dense.shape}")
```

```python
# Get recommendations for specific researcher
def get_grant_recommendations(researcher_id):
   # Load saved components
   researcher_features = np.load('researcher_features.npy')
   grant_features = np.load('grant_features.npy')

   with open('researcher_ids.json', 'r') as f:
       researcher_ids = json.load(f)
   with open('grant_ids.json', 'r') as f:
       grant_ids = json.load(f)

   # Get recommendations
   researcher_idx = researcher_ids.index(researcher_id)
   similarities = cosine_similarity(
       researcher_features[researcher_idx].reshape(1, -1),
       grant_features
   )[0]

   top_indices = similarities.argsort()[-5:][::-1]

   # Get full grant info
   recommended_grants = []
   for idx in top_indices:
       grant_id = grant_ids[idx]
       grant_info = grants_df[grants_df['opportunity_id'] == grant_id].iloc[0]

       recommended_grants.append({
           'grant_id': int(grant_id),
           'similarity_score': float(similarities[idx]),
           'title': grant_info['opportunity_title'],
           'funding_amount': grant_info['estimated_total_funding'],
           'close_date': grant_info['close_date'].strftime('%Y-%m-%d'),
           'agency': grant_info['agency_name'],
           'summary': grant_info['summary_description'][:300] + '...' # Truncated summary
       })

   return recommended_grants

# Test with specific researcher
test_researcher_id = "https://openalex.org/A5046299069"  # Replace with actual ID
recommendations = get_grant_recommendations(test_researcher_id)

for rec in recommendations:
   print(f"\nGrant ID: {rec['grant_id']}")
   print(f"Similarity Score: {rec['similarity_score']:.3f}")
   print(f"Title: {rec['title']}")
   print(f"Funding: ${rec['funding_amount']:,.2f}")
   print(f"Close Date: {rec['close_date']}")
   print(f"Agency: {rec['agency']}")
   print(f"Summary: {rec['summary']}")
```

```python
def evaluate_recommendations():
    # Load components
    researcher_features = np.load('researcher_features.npy')
    grant_features = np.load('grant_features.npy')

    # Calculate full similarity matrix
    full_similarities = cosine_similarity(researcher_features, grant_features)

    # Calculate metrics
    avg_similarity = np.mean(full_similarities)
    max_similarity = np.max(full_similarities)
    coverage = len(np.unique(full_similarities.argmax(axis=1))) / grant_features.shape[0]

    # Distribution of top-5 similarities
    top_5_sims = np.sort(full_similarities, axis=1)[:, -5:]
    avg_top_5 = np.mean(top_5_sims)

    # Category diversity
    unique_agencies = len(grants_df['agency_name'].unique())
    recommended_agencies = len(np.unique(grants_df.iloc[full_similarities.argmax(axis=1)]['agency_name']))
    agency_coverage = recommended_agencies / unique_agencies

    return {
        'average_similarity': avg_similarity,
        'max_similarity': max_similarity,
        'grant_coverage': coverage,
        'average_top5_similarity': avg_top_5,
        'agency_coverage': agency_coverage
    }

metrics = evaluate_recommendations()
print(metrics)
```

```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
import faiss
import pickle
import json

# Load data
researchers_df = pd.read_csv('researchers_clean.csv')
grants_df = pd.read_csv('grants_final_cleaned.csv')

# Fill nulls (same as before)
for col in researchers_df.columns:
    if researchers_df[col].dtype == 'object':
        researchers_df[col] = researchers_df[col].fillna('Unknown')
    else:
        researchers_df[col] = researchers_df[col].fillna(0)

grants_df['close_date'] = pd.to_datetime(grants_df['close_date'], errors='coerce')
grants_df['close_date'] = grants_df['close_date'].fillna(pd.to_datetime('2024-01-01'))

for col in grants_df.columns:
    if grants_df[col].dtype == 'object':
        grants_df[col] = grants_df[col].fillna('Unknown')
    elif pd.api.types.is_numeric_dtype(grants_df[col]):
        grants_df[col] = grants_df[col].fillna(0)
    else:
        grants_df[col] = grants_df[col].fillna('Unknown')

# Create text representations
def create_researcher_text(row):
    return f"""Research focus: {row['concept_1']}, {row['concept_2']}.
              Keywords: {row['top_work_1_keywords']}, {row['top_work_2_keywords']},
              {row['top_work_3_keywords']}"""

def create_grant_text(row):
    return f"""Grant title: {row['opportunity_title']}.
              Categories: {row['funding_categories']}.
              Description: {row['summary_description']}"""

researcher_texts = researchers_df.apply(create_researcher_text, axis=1).tolist()
grant_texts = grants_df.apply(create_grant_text, axis=1).tolist()

# Generate embeddings
model = SentenceTransformer('all-MiniLM-L6-v2')
researcher_embeddings = model.encode(researcher_texts, show_progress_bar=True)
grant_embeddings = model.encode(grant_texts, show_progress_bar=True)

# Build FAISS index
dimension = researcher_embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(grant_embeddings.astype('float32'))

# Save components
np.save('researcher_embeddings.npy', researcher_embeddings)
np.save('grant_embeddings.npy', grant_embeddings)
faiss.write_index(index, 'grants_index.faiss')

researcher_ids = researchers_df['researcher_id'].tolist()
grant_ids = grants_df['opportunity_id'].tolist()

with open('researcher_ids.json', 'w') as f:
    json.dump(researcher_ids, f)
with open('grant_ids.json', 'w') as f:
    json.dump(grant_ids, f)
```

```python
# Updated recommendation system with improvements
from sentence_transformers import SentenceTransformer
import numpy as np
import pandas as pd
import faiss

# Load data
researchers_df = pd.read_csv('researchers_clean.csv')
grants_df = pd.read_csv('grants_final_cleaned.csv')

# Enhanced text representation
def create_researcher_text(row):
   return f"""Expert in {row['concept_1']} and {row['concept_2']}.
             Publications: {row['top_work_1_keywords']}, {row['top_work_2_keywords']}.
             Collaborators: {row['top_collaborator_1']}, {row['top_collaborator_2']}.
             Metrics: h-index {row['h_index']}, citations {row['total_citations']}"""

def create_grant_text(row):
   return f"""Grant: {row['opportunity_title']}.
             Field: {row['funding_categories']}.
             Amount: {row['estimated_total_funding']}.
             Details: {row['summary_description']}"""

# Generate embeddings
model = SentenceTransformer('allenai/scibert_scivocab_uncased')
researcher_texts = researchers_df.apply(create_researcher_text, axis=1).tolist()
grant_texts = grants_df.apply(create_grant_text, axis=1).tolist()

researcher_embeddings = model.encode(researcher_texts, show_progress_bar=True)
grant_embeddings = model.encode(grant_texts, show_progress_bar=True)

# Build FAISS index
dimension = researcher_embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(grant_embeddings.astype('float32'))

def get_filtered_recommendations(researcher_id, min_funding=10000, top_k=5):
   researcher_idx = researcher_ids.index(researcher_id)
   query_vector = researcher_embeddings[researcher_idx].reshape(1, -1)

   D, I = index.search(query_vector.astype('float32'), 20)

   recommendations = []
   for i, idx in enumerate(I[0]):
       grant = grants_df.iloc[idx]
       if grant['estimated_total_funding'] >= min_funding:
           similarity = 1/(1 + D[0][i])
           recommendations.append({
               'grant_id': int(grant['opportunity_id']),
               'similarity': similarity,
               'title': grant['opportunity_title'],
               'funding': float(grant['estimated_total_funding']),
               'agency': grant['agency_name']
           })
           if len(recommendations) == top_k:
               break
   return recommendations

# Test
researcher = researchers_df.iloc[0]
print(f"Testing for: {researcher['researcher_name']} ({researcher['concept_1']})")
recs = get_filtered_recommendations(researcher['researcher_id'], min_funding=50000)
for r in recs:
   print(f"\nScore: {r['similarity']:.3f}")
   print(f"Title: {r['title']}")
   print(f"Agency: {r['agency']}")
   print(f"Funding: ${r['funding']:,.2f}")

# Save components
np.save('researcher_embeddings.npy', researcher_embeddings)
np.save('grant_embeddings.npy', grant_embeddings)
faiss.write_index(index, 'grants_index.faiss')
```
